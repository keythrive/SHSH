###  高校3年生の国語、英語、数学のテスト成績データのサンプリング

-　生成AIへのプロンプト作成

高校３年生１２０名の国語、英語、数学のテストスコアのデータサンプリングを行います。
国語と英語のテストスコアが線形回帰の数式：Yj = Aje * Xe + Bje + eje で与えられます。ここでejeは、誤差の標準偏差で1.2とします。Yjは国語の成績、Xeは英語の成績、Ajeは0.9 Bjeは15とします。

英語と数学のテストスコアが線形回帰の数式：Ye = Aem * Xm + Bem + eem で与えられます。ここでeemは、誤差の標準偏差で2とします。Yeは英語の成績、Xmは数学の成績、Aemは1.3 Bjeは8とします。

数学と国語テストスコアが線形回帰の数式：Ym = Amj * Xj + Bmj + emj で与えられます。ここでemjは、誤差の標準偏差で2.5とします。Ymは数学の成績、Xjは国語の成績、Amjは0.7 Bmjは13とします。

以上の条件で、120名の生徒の成績を正規分布でランダムサンプリングしてください。そのデータは3次元のMatplotでグラフィクスで可視化してください。

まず、各科目のテストスコアの平均と標準偏差を計算します。それぞれの科目について、以下の式を使います。

$$\mu = \frac{\sum_{i=1}^{n} x_i}{n}$$

$$\sigma = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \mu)^2}{n}}$$

ここで、$x_i$は$i$番目の生徒のテストスコア、$n$は生徒の数、$\mu$は平均、$\sigma$は標準偏差です。

次に、正規分布に従って、各科目のテストスコアをランダムサンプリングします。それぞれの科目について、以下の式を使います。

$$x = \mu + \sigma z$$

ここで、$x$はサンプリングされたテストスコア、$z$は標準正規分布からランダムに選ばれた値です。

数学と国語テストスコアが線形回帰の数式：$Y_m = A_{mj} * X_j + B_{mj} + e_{mj}$ で与えられると仮定します。ここで$e_{mj}$は、誤差の標準偏差で2.5とします。$Y_m$は数学の成績、$X_j$は国語の成績、$A_{mj}$は0.7 $B_{mj}$は20とします。
